```
import matplotlib.pyplot as plt
```

```
# サンプルデータ列の読み込み
# AISを連続で受信するような状態
```

```
# データ定義
os = {
    "mmsi": "123456789",
    "latLng": [0, 0.1],
    "sog": 15,
    "cog": 0,
}

ts = {
    "mmsi": "222333444",
    "latLng": [0.3,0.7],
    "sog": 11,
    "cog": 270,
}
```

```


# 初期データのプロット
fig, ax = plt.subplots()
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')
ax.set_title('initial environment')

os_obj, = ax.plot([],[],'o')
ts_obj, = ax.plot([],[],'^')s

os_obj.set_data(os["latLng"])
ts_obj.set_data(ts["latLng"])

plt.show()
```

```
# DCPA, TCPAを計算する関数
def calc_dcpa_tcpa(os, ts):
  dcpa = 1
  tcpa = 2
  return [dcpa, tcpa]

# 他船のDCPA, TCPAが設定値以下なら危険(true)と判定する関数
def simple_alert(os, ts, dcpa_thresh, tcpa_thresh):
  dcpa, tcpa = calc_dcpa_tcpa(os, ts)
  return True if (dcpa < dcpa_thresh and tcpa < tcpa_thresh) else False
```

```
# 実験
print("danger!!!") if simple_alert(os, ts, 2, 3) else print("safe.")
```
